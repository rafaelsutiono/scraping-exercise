```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd


# downloading Metro TV data
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://www.metrotvnews.com/news'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")


article_title = []
bad_program_name = []
alt_img_text = []
bad_description = []


for a in soup.select('h3 a'):
    article_title.append(a.get_text())

for p in soup.select('div.text span'):
    bad_program_name.append(p.get_text())
    uh_program_name = [p.strip() for p in bad_program_name]
    program_name = [p.split(' •')[0] for p in uh_program_name]

imgsrcs = [i['src'] for i in soup.select('div.item img')] # method to extract img src

for alt in soup.select('div.item img'):
    alttext = alt['alt']
    alt_img_text.append(alttext)

for d in soup.select('div.text p'):
    bad_description.append(d.get_text())
    description = [d.replace('\n', '').replace('\r', '') for d in bad_description] # remove \n \r operators
    

print(article_title)
print(program_name)
print(imgsrcs)
print(alt_img_text)
print(description)
```

```python
df = pd.DataFrame(
    {'Article Title': article_title, 'Program Name': program_name, 'Image source': imgsrcs, 'Alt Text': alt_img_text, 'Description': description}
    )

print (df.head())

df.to_csv('MetroTVarticles.csv', index=False)
```
